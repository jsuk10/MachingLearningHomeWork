```python
# 데이터 로드
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
# shape 확인
print(format(X.shape), format(y.shape))
# split 75:25?
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
# split확인
print(format(X_train.shape), format(X_test.shape))
#  k =3 주고 테스트 데이터를 출력하고 성능 확인.
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train,y_train)
print(format(clf.predict(X_test)))
print(format(clf.score(X_test,y_test)))
```

```python
# 위의 암 데이터를 기반으로 knn작성
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,random_state=66)

# 배열 생성
training_accuracy = []
test_accuracy = []

# for으로 1~10까지 작성
neighbors_settings = range(1,11)
for n_neighbors in neighbors_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)
    # 훈련 정확도 저장
    training_accuracy.append(clf.score(X_train,y_train))
    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test,y_test))

plt.plot(neighbors_settings, training_accuracy,label="훈련 정확도")
plt.plot(neighbors_settings, test_accuracy,label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()

```

```python
# 과제
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

count = 0
optimalPoint = 1
difference = []

# for으로 1~10까지 작성
neighbors_settings = range(1,11)
for counts in neighbors_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=counts)
    clf.fit(X_train,y_train)
    # 최적해 탐색
    if((clf.score(X_test,y_test)) < optimalPoint):
        difference.append(clf.score(X_test,y_test))
        count = counts
        optimalPoint = (clf.score(X_test,y_test))

# print("최적해 : "count,"값은" optimalPoint)
# 파이썬에서 도중에 문자를 삽입할 경우 {}을 이용한다 템플릿 문자열 사용가능
# k값의 최적해()
print("테스트 값중 가장 높은 정확도를 가진 최대값은 k 값이 : {} 차이값은 {}".format(count, optimalPoint))

# 그래프를 통해 훈련 정확도와 테스트 정확도의 차이를 알 수 있다.
plt.plot(neighbors_settings, training_accuracy, label="차이값")
plt.ylabel("차이값")
plt.xlabel("n_neighbors")
plt.legend()
```
